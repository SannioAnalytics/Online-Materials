```python
from sksos import SOS
```

```python
detector = SOS(perplexity=10)
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
detector.predict(iris.data)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
```

```python
df.head()
```

```python
df["outlier"] = detector.predict(df.values)
```

```python
df.head()
```

```python
df.sort_values("outlier", ascending=False)
```

```python
from sklearn.decomposition import PCA
```

```python
X = df.drop("outlier", axis = 1).values
X = df.values[:,0:4]
X = iris.data
```

```python
pca = PCA(n_components=2)
X_map = pca.fit_transform(X)
```

```python
df["x"] = X_map[:,0]
df["y"] = X_map[:,1]
```
